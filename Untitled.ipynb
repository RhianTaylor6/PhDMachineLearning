## Deep Learning Methods

```python
# Scaling the data
# Define the scaler 
scaler = StandardScaler().fit(X_train)

# Scale the train set
X_train1 = scaler.transform(X_train)

# Scale the test set
X_test1 = scaler.transform(X_test)
```

```python
scaler
```

```python
model = Sequential()
#start with more units than dimensions and get smaller to 1 output
#to be deep learning it needs more than 1 hidden layer
model.add(Dense(units=30, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
#sigmoid for binary classification, softmax for multi class

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
modelB = Sequential()
#start with more units than dimensions and get smaller to 1 output
#to be deep learning it needs more than 1 hidden layer
modelB.add(Dense(units=30, activation='relu', input_shape=(X_train.shape[1],)))
modelB.add(Dense(units=1, activation='sigmoid'))
#sigmoid for binary classification, softmax for multi class

modelB.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
modelC = Sequential()
#start with more units than dimensions and get smaller to 1 output
#to be deep learning it needs more than 1 hidden layer
modelC.add(Dense(units=30, activation='relu', input_shape=(X_train.shape[1],)))
modelC.add(Dense(units=20, activation='relu'))
modelC.add(Dense(units=1, activation='sigmoid'))
#sigmoid for binary classification, softmax for multi class

modelC.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
modelD = Sequential()
#start with more units than dimensions and get smaller to 1 output
#to be deep learning it needs more than 1 hidden layer
modelD.add(Dense(units=40, activation='relu', input_shape=(X_train.shape[1],)))
modelD.add(Dense(units=30, activation='relu'))
modelD.add(Dense(units=20, activation='relu'))
modelD.add(Dense(units=10, activation='relu'))
modelD.add(Dense(units=1, activation='sigmoid'))
#sigmoid for binary classification, softmax for multi class

modelD.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
#set early stopping monitor so the model stops training to avoid overfitting
early_stopping_monitor = EarlyStopping(patience=3)
#train model
history = model.fit(X_train1, y_train, validation_data=(X_test1,y_test), epochs=30, callbacks=[early_stopping_monitor])
#loss_and_metrics = model.evaluate(X_test1, y_test)
```

```python
historyB = modelB.fit(X_train1, y_train, validation_data=(X_test1,y_test), epochs=30, callbacks=[early_stopping_monitor])

```

```python
historyC = modelC.fit(X_train1, y_train, validation_data=(X_test1,y_test), epochs=30, callbacks=[early_stopping_monitor])
```

```python
historyD = modelD.fit(X_train1, y_train, validation_data=(X_test1,y_test), epochs=30, callbacks=[early_stopping_monitor])
```
